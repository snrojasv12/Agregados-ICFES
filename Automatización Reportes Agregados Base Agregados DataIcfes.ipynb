#### DEPARTAMENTO: 
Escribir el nombre del departamento analizado en el reporte.
En mayúsculas y sin tildes.

#### MUNICIPIO:
Escribir el nombre del municipio principal (ciudad capital) del departamento analizado en el reporte.
En mayúsculas (puede incluir tilde en caso de tener).

#### AÑO:
Escribir el año desde el que se desea hacer el reporte.

```python
DEPARTAMENTO='SAN ANDRES'
MUNICIPIO='SAN ANDRÉS'
AÑO=2017
```

## Tabla Departamentos

```python
s11_cumcap=s11_cumca.drop(s11_cumca[s11_cumca.periodo<AÑO].index)
s11_cumcar=s11_cumcap.drop(s11_cumcap[s11_cumcap.agregacion!='DEPARTAMENTO'].index)
agregados_dptoA=pd.crosstab(index=s11_cumcar.nombre_departamento,columns=s11_cumcar.periodo,dropna=True,values=s11_cumcar.promedio_global,aggfunc='mean',margins=True,margins_name='PROMEDIO')
agregados_dptoA=agregados_dptoA.sort_values(by=['PROMEDIO'],ascending=False)
pd.DataFrame(agregados_dptoA)
```

## Tabla Muncipios

```python
s11_cumcam=s11_cumcap.drop(s11_cumcap[s11_cumcap.agregacion!='MUNICIPIO'].index)
s11_cumcam=s11_cumcam.drop(s11_cumcam[s11_cumcam.nombre_departamento!=DEPARTAMENTO].index)
agregados_muniA=pd.crosstab(index=[s11_cumcam.nombre_departamento,s11_cumcam.nombre_municipio],columns=s11_cumcam.periodo,dropna=True,values=s11_cumcam.promedio_global,aggfunc='mean',margins=True,margins_name='PROMEDIO')
agregados_muniA=agregados_muniA.sort_values(by=['PROMEDIO'],ascending=False)
pd.DataFrame(agregados_muniA)
```

## Tabla Establecimientos Educativos

```python
s11_cumcac=s11_cumcap.drop(s11_cumcap[s11_cumcap.agregacion!='INSTITUCIÓN'].index)
s11_cumcac=s11_cumcac.drop(s11_cumcac[s11_cumcac.nombre_municipio!=MUNICIPIO].index)
s11_cumcac=s11_cumcac.drop(s11_cumcac[s11_cumcac.nombre_departamento!=DEPARTAMENTO].index)
agregados_coleA=pd.crosstab(index=[s11_cumcac.nombre_municipio,s11_cumcac.nombre_establecimiento],columns=s11_cumcac.periodo,dropna=True,values=s11_cumcac.promedio_global,aggfunc='mean',margins=True,margins_name='PROMEDIO')
pd.DataFrame(agregados_coleA)
```

### Exportar Tablas a Excel

```python
with pd.ExcelWriter('Z:/Datos/DataIcfes/Resultados Saber11/REPORTE '+MUNICIPIO+'-'+DEPARTAMENTO+'.xlsx') as writer:
    agregados_dptoA.to_excel(writer,sheet_name=DEPARTAMENTO+' - PAÍS',float_format="%.0f",index=True,index_label='DEPARTAMENTO')
    agregados_muniA.to_excel(writer,sheet_name='MUNICIPIOS - '+DEPARTAMENTO,float_format="%.0f",index=True,index_label=['DEPARTAMENTO','MUNICIPIO'])
    agregados_coleA.to_excel(writer,sheet_name='EE '+MUNICIPIO,float_format="%.0f",index=True,index_label=['MUNICIPIO','NOMBRE ESTABLECIMIENTO EDUCATIVO'])
```
