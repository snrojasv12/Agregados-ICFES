```python
DEPARTAMENTO='CHOCO'
MUNICIPIO='QUIBDÓ'
AÑO=2018
```

## Tabla Departamentos

```python
s11_cumcar=s11_cumca.drop(s11_cumca[s11_cumca.PERIODO<AÑO].index)
agregados_dptoA=pd.crosstab(index=s11_cumcar.ESTU_DEPTO_RESIDE,columns=s11_cumcar.PERIODO,dropna=True,values=s11_cumcar.PUNT_GLOBAL,aggfunc='mean',margins=True,margins_name='PROMEDIO',)
#revisar variable dpto pues hay 3 tipos
agregados_dptoA=agregados_dptoA.sort_values(by=['PROMEDIO'],ascending=False)
pd.DataFrame(agregados_dptoA)
```

## Tabla Muncipios

```python
s11_cumcam=s11_cumcar.drop(s11_cumcar[s11_cumcar.ESTU_DEPTO_RESIDE!=DEPARTAMENTO].index)
#revisar variable municipio pues hay 3 tipos
agregados_muniA=pd.crosstab(index=[s11_cumcam.ESTU_DEPTO_RESIDE,s11_cumcam.ESTU_MCPIO_RESIDE],columns=s11_cumcam.PERIODO,dropna=True,values=s11_cumcam.PUNT_GLOBAL,aggfunc='mean',margins=True,margins_name='PROMEDIO',)
agregados_muniA=agregados_muniA.sort_values(by=['PROMEDIO'],ascending=False)
pd.DataFrame(agregados_muniA)
```

## Tabla Establecimientos Educativos

```python
s11_cumcac=s11_cumcam.drop(s11_cumcam[s11_cumcam.ESTU_MCPIO_RESIDE!=MUNICIPIO].index)
agregados_coleA=pd.crosstab(index=[s11_cumcam.ESTU_MCPIO_RESIDE,s11_cumcam.COLE_NOMBRE_ESTABLECIMIENTO],columns=s11_cumcac.PERIODO,dropna=True,values=s11_cumcac.PUNT_GLOBAL,aggfunc='mean',margins=True,margins_name='PROMEDIO',)
pd.DataFrame(agregados_coleA)
```

### Exportar Tablas a Excel

```python
with pd.ExcelWriter('Z:/Datos/DataIcfes/Resultados Saber11/REPORTE '+DEPARTAMENTO+'.xlsx') as writer:
    agregados_dptoA.to_excel(writer,sheet_name=DEPARTAMENTO+' - PAÍS',float_format="%.0f",index=True,index_label='DEPARTAMENTO')
    agregados_muniA.to_excel(writer,sheet_name='MUNICIPIOS - '+DEPARTAMENTO,float_format="%.0f",index=True,index_label=['DEPARTAMENTO','MUNICIPIO'])
    agregados_coleA.to_excel(writer,sheet_name='EE '+MUNICIPIO,float_format="%.0f",index=True,index_label=['MUNICIPIO','NOMBRE ESTABLECIMIENTO EDUCATIVO'])
```
